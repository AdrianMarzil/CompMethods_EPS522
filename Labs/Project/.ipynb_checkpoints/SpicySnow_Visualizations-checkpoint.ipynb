## Plotting Spicy_Snow NetCDF Outputs

```python
import numpy as np
import pandas as pd
import xarray as xr

import glob
import geopandas
import rioxarray
import re

import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker

from datetime import datetime
from cartopy.util import add_cyclic_point
from shapely.geometry import mapping

import netCDF4 as nc
from netCDF4 import Dataset

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.colors
from mpl_toolkits.basemap import Basemap
```

### Import SNOTEL Data

```python
# Import SNOTEL Excel File
SNOTEL = pd.ExcelFile("/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak Ranch/SNOTEL DATA/BPR_SNOTELs_2016-2023.xlsx")
cham = pd.read_excel(SNOTEL, 'Chamita')
cumb = pd.read_excel(SNOTEL, 'LilyPond')
lily = pd.read_excel(SNOTEL, 'CumbresTrestle')
wolf = pd.read_excel(SNOTEL, 'WolfCreek')

# Convert inches to meters
cham_m = cham['SnowDepth[in]'] * 0.0254
cumb_m = cumb['SnowDepth[in]'] * 0.0254
lily_m = lily['SnowDepth[in]'] * 0.0254
wolf_m = wolf['SnowDepth[in]'] * 0.0254

Date = pd.to_datetime(cumb['Date'])
#Date
```

### Plot SNOTEL Data

```python
# Plot SNOTEL data for years 2016 - 2023
fig = plt.figure(figsize = (10, 5), dpi=150)
ax = fig.add_subplot()

ax.plot(Date, lily_m)
ax.plot(Date, cumb_m)
ax.plot(Date, cham_m)
ax.plot(Date, wolf_m)
ax.set_xlabel('Date')
ax.set_ylabel('Snow Depth [m]')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))
ax.set_xlim([datetime(2015, 12, 1), datetime(2023, 6, 1)])
ax.set_ylim(0, 4)
ax.grid(linestyle=':', linewidth='0.5', color='grey')

ax.set_title("SNOTEL Snow Depth: 2016-2023", fontweight="bold")

ax.legend([ 'Lily Pond (3,374 m)', 'Cumbres Trestle (3,059 m)', 'Chamita (2,555 m)', 'Wolf Creek Summit (3,340 m)'])
```

### Import Catchment Shapefiles

```python
# Set File path, Filename and CRS for shapefile and rename to 'SF'
SFpath = '/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak Ranch/BPR GIS/GIS FILES/Basins & Catchments/Chama Basins/'
SFfname = 'NavajoRiver_Basin_toOsoDiv.shp'
SF = geopandas.read_file(SFpath+SFfname, crs="epsg:4326")
```

### Plot SpicySnow Retrievals

```python
#count = len(ds.variables['snow_depth'])
count = len(ds.variables['time'])

for i in range(count):
    
    # Set CRS for dataset
    ds.rio.write_crs('epsg:4326 ', inplace=True)
    
    # Clip with Shapefile 'SF'
    CLP = ds.rio.clip(SF.geometry.apply(mapping), SF.crs)
    
    fig = plt.subplots(figsize = (7,5), dpi=100)
    snd = CLP['snow_depth'][:, :, i]
    lat = CLP['y']
    long = CLP['x']
    #date = CLP.time.values[i]
    date = np.datetime_as_string(CLP.time.values[i], unit='D')
    
    # Plot Snow_Depth
    levels = np.linspace(0, 4.5, 19)
    plt.contourf(long,lat, snd, levels, cmap="Blues")
    plt.colorbar(label='Snow Depth [m]')
    plt.title('SpicySnow Snow Depth: '+str(date), fontweight="bold")
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    plt.show()
```

### Plot Wet Snow Mask (SpicySnow)

```python
#count = len(ds.variables['snow_depth'])
count = len(ds.variables['time'])

for i in range(count):
    
    # Set CRS for dataset
    ds.rio.write_crs('epsg:4326 ', inplace=True)
    
    # Clip with Shapefile 'SF'
    CLP = ds.rio.clip(SF.geometry.apply(mapping), SF.crs)
    
    fig = plt.subplots(figsize = (7,5), dpi=100)
    snd = CLP['wet_snow'][i, :, :]
    lat = CLP['y']
    long = CLP['x']
    #date = CLP.time.values[i]
    date = np.datetime_as_string(CLP.time.values[i], unit='D')
    
    # Plot Snow_Depth
    levels = np.linspace(0, 1, 3)
    cmap = matplotlib.colors.ListedColormap(['white', 'firebrick'])
    plt.contourf(long, lat, snd, levels, cmap=cmap)
    plt.colorbar(label='Wet Snow')
    plt.title('SpicySnow Wet Snow: '+str(date), fontweight="bold")
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    
    plt.show()
```

###  SpicySnow: Calculate Mean Snow Depth For Catchment

```python
# Set File Path and Filename for netCDF files
#path='/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak/SND_NC/'
path='/Users/Adrian/Desktop/SpicySnow Data/'
fname='BPR_snd2023.nc'

# Set File path, Filename and CRS for shapefile and rename to 'SF'
#SFpath = '/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak Ranch/BPR GIS/GIS FILES/Basins & Catchments/Chama Basins/'
#SFfname = 'NavajoRiver_Basin_toOsoDiv.shp'
#SFpath = '/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak Ranch/BPR GIS/GIS FILES/CatchmentContours/'
#SFfname = 'NavajoBasin_AreaBlw2700m.shp'
SFpath = '/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak Ranch/BPR GIS/GIS FILES/Survey Measurement Locations/SurveyAreaBoxes/'
SFfname = 'JuliesMeadowBox.shp'
SF = geopandas.read_file(SFpath+SFfname, crs="epsg:4326")

ds = xr.open_dataset(path+fname)

# Create new lists for mean snow depth
meanS1snd = []

count = len(ds.variables['time'])

# Plot Mean for netCDF files
for i in range (count):
    
    # Set CRS for dataset
    ds.rio.write_crs('epsg:4326 ', inplace=True)

    # Clip with Shapefile 'SF'
    CLP = ds.rio.clip(SF.geometry.apply(mapping), SF.crs, drop=True)
    snd = CLP['snow_depth'][:,:,i]
    time = CLP['time']
    #snd = snd.drop(['time'])

    # Calculate snow depth mean and populate list
    meanS1snd.append(np.mean(snd))
    
    # Drop NaN's from snow depth
    #meanS1snd_nonan = meanS1snd.dropna()
```

###  Lievens Data: Calculate Mean Snow Depth For Catchment

```python
# Set File Path and Filename for netCDF files
#path='/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak/SND_NC/'
path='/Users/Adrian/Desktop/LievensS1Data/S1 SanJuan Clipped Data/'
fname='snd_2021*.nc'

# Set File path, Filename and CRS for shapefile and rename to 'SF'
#SFpath = '/Users/Adrian/OneDrive - University of New Mexico/SNOW RESEARCH/SNOW DATA/Banded Peak Ranch/BPR GIS/GIS FILES/Basins & Catchments/Chama Basins/'
#SFfname = 'NavajoRiver_Basin_toOsoDiv.shp'
#SF = geopandas.read_file(SFpath+SFfname, crs="epsg:4326")

# Create new lists
MeanLS1snd = []
S1Date = []

# Plot Mean for netCDF files
for file in glob.glob(path+fname):
    
    #fig = plt.subplots(figsize = (10,6))
    ds = xr.open_dataset(file)
    
    # Set CRS for dataset
    ds.rio.write_crs('epsg:4326 ', inplace=True)
    
    # Search File Path for Date, Rename as 'date', Change to DD-MM-YYYY format
    match_str = re.search(r'\d{4}\d{2}\d{2}', file)
    datetime_obj = datetime.strptime(match_str.group(), '%Y%m%d').date()
    #date = datetime_obj.strftime('%d-%m-%Y')
   
    
    # Add date to S1Date list
    S1Date.append(datetime_obj) 

    
    # Clip with Shapefile 'SF'
    CLP = ds.rio.clip(SF.geometry.apply(mapping), SF.crs, drop=True)
    
    # Calculate snd Mean
    MeanLS1snd.append(np.nanmean(CLP['snd']))
```

```python
# Create DataFrame from lists and drop NaN values
SnowDF = pd.DataFrame(np.column_stack([S1Date, MeanLS1snd]), columns=['Date', 'MeanS1Depth'])
SnowDF.sort_values(by='Date', ascending = True, inplace = True)
SnowDF_nonan = SnowDF.dropna()
```
